```python
import pandas as pd
import numpy as np

from ast import literal_eval
```

```python
# Read csv file, drop unwanted columns and rows
df = pd.read_csv('./colonel_blotto.csv', header=0)
df.drop(['Unnamed: 22', 'Unnamed: 23'],axis=1, inplace=True)
df.drop(21, axis=0, inplace=True)

# Replace strings in P1 column by tuples
df['P1'] = [literal_eval(x) for x in df['P1']]
```

```python
'''
Given two tuples, compute the utility in the form of a tuple.
'''
def getUtility(tuple1, tuple2):
    # Sanity check
    if (len(tuple1) != len(tuple2)):
        print("Your inputs don't match dumbass")
        return -1
    
    win1, win2 = 0, 0
    for i in range(len(tuple1)):
        if (tuple1[i] > tuple2[i]):
            win1+=1
        elif (tuple1[i] < tuple2[i]):
            win2+=1
        else:
            continue
    return (win1, win2)
```

```python
# Populate the dataframe, iterate through df to compute utility
rows, cols = df.shape[0], df.shape[1]
df.iloc[0][1]
colname = df.columns[2]
colname_tuple = literal_eval(colname)

for r in range(1, rows):
    for c in range(1, cols):
        tuple1 = df.iloc[r][0]
        tuple2 = literal_eval(df.columns[c])
        df.iloc[r][c] = getUtility(tuple1, tuple2)
```
